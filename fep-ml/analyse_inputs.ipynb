## Explore Covid-Moonshot/F@H data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pd.options.display.max_rows = 20
cols_of_interest = ["SMILES", "canonical_CID", "r_avg_IC50", "f_avg_IC50", "trypsin_IC50"]
activity_data = pd.read_csv("input_data/activity_data.csv", usecols=cols_of_interest)
activity_data
```

```python
#Compute pIC50 as rapidfire, else fluorescence, else trypsin, else nan:
pIC50s = []

for idx, row in activity_data.iterrows():
    r_ic50 = row[2]
    f_ic50 = row[3]
    t_ic50 = row[4]
    
    if not row.isna()["r_avg_IC50"]:
        pIC50 = r_ic50
    elif not row.isna()["f_avg_IC50"]:
        pIC50 = f_ic50
    elif not row.isna()["trypsin_IC50"]:
        pIC50 = t_ic50
    else:
        pIC50 = False
    if pIC50:
        pIC50s.append([row[0], pIC50])

print(len(pIC50s))
```
